```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_200.txt",delimiter=" ",header=None, names=("date","time","ir","lidar"))

d = data.loc[:,["ir","lidar"]] #光センサとlidarのデータだけにする"
```

```python
sns.jointplot(d["ir"],d["lidar"],d,kind="kde") #度数分布を描画
d.cov() #分散、共分散を計算
```

```python
import numpy as np
from scipy.stats import multivariate_normal

x,y = np.mgrid[280:340, 190:230]
pos = np.empty(x.shape + (2,))
pos[:,:,0] = x
pos[:,:,1] = y

irlidar = multivariate_normal(mean=d.mean().values.T, cov=d.cov().values) #2次元ガウス分布を作る
cont = plt.contour(x,y,irlidar.pdf(pos)) #密度の値の等高線を作る
cont.clabel(fmt="%1.1e") #等高線の値の文字列を入れる
plt.show()
```

共分散が -13.4...でガウス分布は右肩下がり

光センサ値とLidarセンサ値は、片方が平均値より大きい場合、もう片方は平均値より小さいことがわかる
