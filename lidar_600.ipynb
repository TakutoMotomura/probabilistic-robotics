```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt",delimiter=" ",header=None,names=("data","time","ir","lidar"))

data["lidar"].hist(bins=max(data["lidar"]) - min(data["lidar"]),align="left")
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"] = [e//10000 for e in data.time]
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = {i:d.lidar.get_group(i).value_counts().sort_index() for i in range(24)} #時間ごとにデータフレームを作成
# 時間 i におけるセンサ値の個数を数え、センサ値順に並べてデータフレームを返す

freqs = pd.concat(each_hour,axis=1) #concatで連結
# 24個のデータフレームを1個に連結

freqs = freqs.fillna(0) #欠損は0で埋める
probs = freqs/len(data) #頻度を確率で割る

probs
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
sns.jointplot(data["hour"],data["lidar"],data,kind="kde")
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum()) #時間（辞書のキー）ごとに確率を合計
p_t.plot()
p_t.transpose()
```

```python
p_t.sum() #1になる
```

```python
p_z = pd.DataFrame(probs.transpose().sum()) #センサ値ごとに確率を合計
p_z.plot()
p_z.transpose()
```

```python
p_z.sum() #1になる
```

```python
cond_z_t = probs/p_t[0] #列（時間）ごとにP(t)で割るとP(x|t)となる
cond_z_t
```

```python
# P(z|t=6)
(cond_z_t[6]).plot.bar(color="blue",alpha=0.5)
# P(z|t=14)
(cond_z_t[14]).plot.bar(color="orange",alpha=0.5)
plt.show()
```
